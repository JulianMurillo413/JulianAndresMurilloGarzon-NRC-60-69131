```python
class Nodo:
    def __init__(self, Nombre, Apellido, Asignatura, Programa, NRC, ID):
        self.Nombre = Nombre
        self.Apellido = Apellido
        self.Asignatura = Asignatura
        self.Programa = Programa
        self.NRC = NRC
        self.ID = ID
        self.siguiente = None

class ListaSimple:
    def __init__(self):
        self.cabeza = None


    def insertar(self, Nombre, Apellido, Asignatura, Programa, NRC, ID):
        nuevo_nodo = Nodo(Nombre, Apellido, Asignatura, Programa, NRC, ID)
        if not self.cabeza:
            self.cabeza = nuevo_nodo
        else:
            actual = self.cabeza
            while actual.siguiente:
                actual = actual.siguiente
            actual.siguiente = nuevo_nodo


    def mostrar(self):
        if not self.cabeza:
            print("La lista está vacía.")
            return
        actual = self.cabeza
        while actual:
            print(f"Nombre: {actual.Nombre}, Apellido: {actual.Apellido}, Asignatura: {actual.Asignatura}, Programa: {actual.Programa}, NRC: {actual.NRC}, ID: {actual.ID}")
            actual = actual.siguiente
        print("None")


lista = ListaSimple()


lista.insertar('Laura', 'Perez', 'Matemáticas', 'Ingeniería', '1234', '001')
lista.insertar('Pedro', 'Picapiedra', 'Historia', 'Humanidades', '5678', '002')


lista.mostrar()


```

```python
class Item:
    def __init__(self, nombre, cantidad, descripcion):
        self.nombre = nombre
        self.cantidad = cantidad
        self.descripcion = descripcion
        self.siguiente = None

class Inventario:
    def __init__(self):
        self.cabeza = None


    def insertar(self, nombre, cantidad, descripcion):
        nuevo_item = Item(nombre, cantidad, descripcion)
        if not self.cabeza:
            self.cabeza = nuevo_item
        else:
            actual = self.cabeza
            while actual.siguiente:
                actual = actual.siguiente
            actual.siguiente = nuevo_item


    def mostrar(self):
        if not self.cabeza:
            print("El inventario está vacío.")
            return
        actual = self.cabeza
        while actual:
            print(f"Nombre: {actual.nombre}, Cantidad: {actual.cantidad}, Descripción: {actual.descripcion}")
            actual = actual.siguiente
        print("None")


inventario_aula = Inventario()


inventario_aula.insertar('Escritorio', 20, 'Escritorio de madera')
inventario_aula.insertar('Computador', 15, 'Computador de escritorio con Windows')
inventario_aula.insertar('Mouse', 15, 'Mouse óptico')
inventario_aula.insertar('Teclado', 15, 'Teclado mecánico')
inventario_aula.insertar('Silla', 20, 'Silla ergonómica')
inventario_aula.insertar('Tablero', 2, 'Tablero blanco')
inventario_aula.insertar('TV', 1, 'Televisor de 50 pulgadas')


inventario_aula.mostrar()
```

```python
class Famoso:
    def __init__(self, famoso, nombre, cantidad_tiktok_subidos):
        self.famoso = famoso
        self.nombre = nombre
        self.cantidad_tiktok_subidos = cantidad_tiktok_subidos
        self.siguiente = None

class ListaFamosos:
    def __init__(self):
        self.cabeza = None


    def insertar(self, famoso, nombre, cantidad_tiktok_subidos):
        nuevo_famoso = Famoso(famoso, nombre, cantidad_tiktok_subidos)
        if not self.cabeza:
            self.cabeza = nuevo_famoso
        else:
            actual = self.cabeza
            while actual.siguiente:
                actual = actual.siguiente
            actual.siguiente = nuevo_famoso


    def mostrar(self):
        if not self.cabeza:
            print("La lista de famosos está vacía.")
            return
        actual = self.cabeza
        while actual:
            famoso_texto = "Sí" if actual.famoso else "No"
            print(f"Famoso: {famoso_texto}, Nombre: {actual.nombre}, Cantidad de TikToks subidos: {actual.cantidad_tiktok_subidos}")
            actual = actual.siguiente
        print("None")

lista_famosos = ListaFamosos()


lista_famosos.insertar(True, 'Rubius', 150)
lista_famosos.insertar(True, 'Mia Khalifa', 120)
lista_famosos.insertar(False, 'Danielito_58', 5)
lista_famosos.insertar(True, 'El Xokas', 50)
lista_famosos.insertar(False, 'Andresito_gammer', 10)


lista_famosos.mostrar()
```
